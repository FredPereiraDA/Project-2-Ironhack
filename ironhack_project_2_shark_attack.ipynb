```python
srk_cl = srk_atk.drop(columns=['Unnamed: 22', 'Unnamed: 23', 'href formula', 'href', 'Injury', 'Fatal (Y/N)', 'Location', 'Species ', 'Area', 'pdf', 'Time', 'Investigator or Source', 'Case Number.1', 'Case Number.2', 'original order'])
srk_cl
```

```python
df = srk_cl.dropna(subset=['Year'])
df
```

```python
df1 = df['Year'].astype(int)
df1
```

```python
df['Year'].unique()
```

```python
num_attacks_country = srk_cl["Country"].value_counts().nlargest(9)
num_attacks_country.sort_values(ascending=True, inplace=True)
num_attacks_country.plot.barh(color="royalblue")

plt.title("Unprovoked Shark Attacks by Country (1967-2017)")
plt.xlabel("Attacks")

plt.grid(axis="x")
plt.tight_layout()
plt.savefig("AttacksByCountry.png")
```

```python
num_attacks_country = srk_cl["Activity"].value_counts().nlargest(9)
num_attacks_country.sort_values(ascending=True, inplace=True)
num_attacks_country.plot.barh(color="darkgreen")

plt.title("Unprovoked Shark Attacks by Acitivity (1967-2017)")
plt.xlabel("Attacks")

plt.grid(axis="x")
plt.tight_layout()
plt.savefig("AttacksByYear.png")
```
